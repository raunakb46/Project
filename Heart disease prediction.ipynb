# Using the dataset in model

```python
# Splitting the dataset into features and target

x = df.drop(columns = 'target', axis = 1)
y = df['target']
```

```python
print (x)
```

```python
print (y)
```

```python
# Splitting the data for training and testing

from sklearn.model_selection import train_test_split

# stratify = y --> Distributes the label evenly within training and testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 2)
```

```python
print (x.shape, x_train.shape, x_test.shape)
```

### Using Logistic regression

```python
# Training the model

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train)
```

#### Evaluation

```python
# For training data

from sklearn.metrics import accuracy_score

train_pred = lr.predict(x_train)
acc_train = accuracy_score(y_train, train_pred)
print ('Accuracy from training data:', acc_train)
```

```python
# For testing data

test_pred = lr.predict(x_test)
acc_test = accuracy_score(y_test, test_pred)
print ('Accuracy from testing data:', acc_test)
```

* As the difference between accuracy from training and testing data is less, the model is `good`.

# Predictive system

```python
# Providing input to the system
inp = (56, 1, 1, 130, 221, 0, 0, 163, 0, 0, 2, 0, 3)

# Converting input to numpy array
inp_arr = np.asarray(inp)

# Reshaping the array as it should take 1 row as input and should not expect the no. of rows used for training
inp_reshape = inp_arr.reshape(1, -1)

prediction = lr.predict(inp_reshape)
if (prediction == 1):
    print ('Person has heart disease')
else:
    print ('Person is healthy')
```
