# Handling the imbalanced dataset

```python
# Authentic contains all the details of original transactions
# Fraud contains all the details of a a fraud transaction

authentic = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
print (authentic.shape)
print (fraud.shape)
```

```python
authentic.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Comparing authentic and fraud transactions

df.groupby('Class').mean()
```

* Here, we observe, `mean amount of transaction in case of fraud is higher` than authentic transaction. 

### Under sampling

```python
# We need to build a dataset which has similar distribution of fraud and authentic transaction in order to balance it
# As we have 492 fraud transaction, creating a sample from random 492 transactions in 'authentic'.

authentic_sam = authentic.sample(n = 492)
```

```python
# Joining this sample with fraud to create a new and balanced dataset

balance_df = pd.concat([authentic_sam, fraud], axis = 0)
```

```python
balance_df.head()
```

```python
balance_df.tail()
```

```python
balance_df['Class'].value_counts()
```

```python
balance_df.groupby('Class').mean()
```

* The mean value of all the features of class 0 and 1 are not varied much, hence the new dataframe is fine.

# Using the new dataset in model

```python
# Spltting the data into features and target

x = balance_df.drop(columns = 'Class', axis = 1)
y = balance_df['Class']
```

```python
print (x)
```

```python
print (y)
```

```python
# Splitting the data for training and testing

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# stratify = 1 is used to distribute the class evenly in training and testing set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 2)
```

### Using Logistic regression

```python
# Training the model

from sklearn.linear_model import LogisticRegression

log = LogisticRegression()
log.fit(x_train, y_train)
```

#### Evaluation

```python
# Accuracy of training set

train_pred = log.predict(x_train)
acc_train = accuracy_score(train_pred, y_train)
print ('Accuracy of training data:', acc_train * 100, '%')
```

```python
# Accuracy of testing set

test_pred = log.predict(x_test)
acc_test = accuracy_score(test_pred, y_test)
print ('Accuracy of training data:', acc_test * 100, '%')
```

* As the difference between accuracy of training and testing data is very less, the model has a `good accuracy`.
