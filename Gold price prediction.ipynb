# Using dataset in model

```python
# Splitting the dataset into fetures and label

x = df.drop(['Date', 'GLD'], axis = 1)
y = df['GLD']
```

```python
x
```

```python
y
```

```python
# Splitting data for training and testing

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
```

### Using Random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Training the model

rfr = RandomForestRegressor(n_estimators = 100)
rfr.fit(x_train, y_train)
```

#### Evaluation

```python
# For training data

from sklearn import metrics
from sklearn.metrics import r2_score

train_pred = rfr.predict(x_train)
acc_train = metrics.r2_score(y_train, train_pred)
print ('Accuracy score for training data:', acc_train)
```

```python
# For testing data

test_pred = rfr.predict(x_test)
acc_test = metrics.r2_score(y_test, test_pred)
print ('Accuracy score for training data:', acc_test)
```

* As the difference between r2 score of training and testing data is less, the model is `good`.

#### Comparing the actual and predicted value on plot

```python
y_list = list(y_test)
```

```python
plt.plot(y_list, color = 'green', label = 'Actual values')
plt.plot(test_pred, color = 'red', label = 'Predicted values')
plt.title('Actual value vs. Predicted value')
plt.ylabel('Gold price')
plt.xlabel('Values')
plt.legend()
plt.show()
```

From the plot -
* Predicted values are very close to actual values.

# Predictive system

```python
# Input for making prediction
inp = (1310.5, 70.550003, 15.902, 1.464794)

# Converting input to array
inp_arr = np.asarray(inp)

# Reshaping the array
inp_reshape = inp_arr.reshape(1, -1)

# Making prediction
inp_pred = rfr.predict(inp_reshape)
```
